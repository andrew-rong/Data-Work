```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

# FR is the fixed rate that will be set between the 2 parties
# Variable rate won't be inputted but randomly generated
# Time is represented in years, but payments will be made on a monthly basis.


```python
class swap:
    # Initializing Variables
    def __init__(self, Asset_value, fixed_rate, years):
        self.value = Asset_value
        self.FR = fixed_rate
        self.t = years * 12
        self.timeline = np.arange(0,self.t)

    # 1 simulation
    def one_sim(self):
        x = np.zeros(self.t)
        for i in range(0, self.t):
            x[i] = self.value * (self.FR + np.random.normal(0, 0.003))
        return x

    # Plotting 1 simulation in comparison
    def plot_one_sim(self):
        y = np.zeros(self.t)

        for i in range(0, self.t):
            y[i] = self.value * self.FR
        plt.figure()
        plt.title("Swap")
        plt.xlabel("Time In Months)")
        plt.ylabel("Amount of Interest Paid")
        plt.plot(self.timeline, self.one_sim(), label="Variable")
        plt.plot(self.timeline, y, label="Fixed")
        plt.legend()
        plt.show()
        print(self.one_sim().sum(), y.sum())

    # A (n simulations) x (time) array containing the different potential outcomes.
    def sim_n_times(self, n):
        a = np.zeros((n, self.t))

        for i in range(n):
            a[i] = self.one_sim()
        return a
    
    # Plotting n simulations
    def plot_n_sims(self, n):
        sims = self.sim_n_times(n)
        for i in range(n):
            plt.plot(self.timeline, sims[i])
            plt.title(f"Swap Simulation {n} Times")
            plt.xlabel("Time In Months)")
            plt.ylabel("Amount of Interest Paid")
```

```python
x = swap(100000, 0.05, 6)
print(x.one_sim())
x.plot_one_sim()
print(x.sim_n_times(2))
x.plot_n_sims(100)
```
