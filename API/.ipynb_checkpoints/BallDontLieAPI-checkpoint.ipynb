```python
response_teams = requests.get("https://www.balldontlie.io/api/v1/teams").json()
response_teams['data'][0]
```

```python
def find_team_id(team):
    for city in response_teams["data"]:
        if city["city"] == team:
            return city["id"]
    
```

```python
find_team_id("Los Angeles")
```
